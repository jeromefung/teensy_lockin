```python
def plotting(x, y, y_label, plot_title):
    plt.figure(figsize=(10,10))
    plt.plot(x,y)
    plt.xlabel("t (s)", fontsize=20)
    plt.ylabel(y_label, fontsize=20)
    plt.title(plot_title, fontsize=25)
    plt.show()
    return None
```
