```python
#import needed libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#set constants
samp_rate = 10000
path = "C:\\Users\\chris\\OneDrive\\Documents\\College\\Spring 2021\\teensy\\teensy_lockin\\teensyGUI\\testResults\\"
```

```python
def plotting(x, y, y_label, plot_title):
    plt.figure(figsize=(10,10))
    plt.plot(x,y)
    plt.xlabel("t (s)", fontsize=20)
    plt.ylabel(y_label, fontsize=20)
    plt.title(plot_title, fontsize=25)
    plt.show()
    return None
```

```python
def analyze(filename):
    data = pd.read_csv(filename, header=0)
    
    t = np.array(data["Samp"])/samp_rate
    signal = np.array(data["Signal"])*3.3/4096
    I = np.array(data["I"])*3.3/4096
    Q = np.array(data["Q"])*3.3/4096
    amp = np.array(data["R"])*2*3.3/4096
    phi = np.array(data["Phi"])
    
    signal_bound = 250
    plotting(t[:signal_bound], signal[:signal_bound], "Voltage (V)", "Measured Signal")
    plotting(t, I, "Voltage (V)", "In-Phase")
    plotting(t, Q, "Voltage (V)", "Out-of-Phase")
    plotting(t, amp, "Voltage (V)", "Measured Amplitude")
    plotting(t, phi, "Phase (Rad)", "Measured Phase")
    return None
```
