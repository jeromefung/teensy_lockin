```python
#starting with correct pin orientation and internal lock in with ref freq of 10000
data_selected = data["Correct pin orientation, Internal"] #only get the signal
data_selected = data_selected[1:] #get rid of label
data_selected = np.array(data_selected) #convert to numpy array
data_selected = data_selected * 3.3 / 4096 #convert to voltages
```

```python
#other constants
samp_rate = 10000
ref_freq = 1000
cutoff_freq = 1
t = np.linspace(0, len(data_selected), len(data_selected))
```

```python
def filter_signal2(x, a, b):
    '''
    As in filter_signal, assume circular boundary condition.
    
    Parameters
    ----------
    x : list 
        Input signal
    a : list
        Filter coefficients. Assume a[0] is the coefficient a0, etc
    b : list
        Filter coefficients. Assume b[1] is b1, etc; keep b[0] = 0.
        
    Returns
    -------
    y : list
        Output (for testing only). In practice, on Teensy, dump output to serial monitor.
        
    Notes
    -----
    Because this is a recursive filter, the input and output values need to be padded at the beginning.
    Currently, pad input circularly (e.g., x[-1]), and assume output y is initially 0.
    It might eventually be better to pad with the initial value of the input??
    '''
    
    n_pts = len(x)
    
    # Preallocate output list -- just for testing here.
    # Could get rid of this list by just printing values as they are calculated.
    y = [0] * n_pts
    
    n_coeffs = len(a) # this is n_poles + 1
    
    # code needs to store some y values in register y_reg, but only n_poles of them
    # but to simplify rest of code, have y[0] be 0 (do nothing), so y_reg is of length n_coeffs
    y_reg = [0] * (n_coeffs)
    # y_reg[0] is 0, y_reg[1] is y[n-1], y-reg[2] is y[n-2], etc.
    # To not initialize with 0, we'd set y_reg to other values here (e.g., first value of input)
    
    # Main loop over the input signal
    for n in range(n_pts):
        yn = 0 # initialize the output
        # Loop over coefficients
        for coeff_ctr in range(n_coeffs):
            yn = yn + a[coeff_ctr] * x[n - coeff_ctr] + b[coeff_ctr] * y_reg[coeff_ctr]
        # Write y to output -- on Teensy we'd dump to serial monitor instead
        y[n] = yn
        # Update register by shifting y_reg[n] to y_reg[n+1]
        for coeff_ctr in range(n_coeffs - 1, 0, -1):
            # need to do the shift backwards
            # special case: updating y_reg[1]
            if coeff_ctr == 1:
                y_reg[coeff_ctr] = yn
            else: # the normal case, shift elements
                y_reg[coeff_ctr] = y_reg[coeff_ctr - 1]
    
    # only for testing purposes -- actual code on Teensy would dump output to serial monitor
    return y
        
```

```python
#put through teensy algorithm
ref_sine = np.sin(2*np.pi*ref_freq*t/samp_rate)
ref_cos = np.cos(2*np.pi*ref_freq*t/samp_rate)

#mix the signals
I = ref_sine * data_selected
Q = ref_cos * data_selected

#make coefficients
x = np.exp(-2*np.pi*cutoff_freq/samp_rate)
a = [1 - x, 0]
b = [0, x]

#use Jerome's filter
I_out = filter_signal2(I, a, b)
Q_out = filter_signal2(Q, a, b)

#calculate amplitude and phase
R = np.sqrt(np.array(I_out)*np.array(I_out) + np.array(Q_out)*np.array(Q_out))
phi = np.arctan2(Q_out,I_out)

#plot everything
plt.figure(figsize=(10,10))
plt.plot(t/samp_rate, 2*R)
plt.xlabel("t")
plt.ylabel("V")
plt.title("Amplitude")
plt.show()

plt.figure(figsize=(10,10))
plt.plot(t/samp_rate, phi)
plt.xlabel("t")
plt.ylabel("radians")
plt.title("Phase Difference")
plt.show()
```
